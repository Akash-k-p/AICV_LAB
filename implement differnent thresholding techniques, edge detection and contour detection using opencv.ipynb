```python
import cv2

image =  cv2.imread('img1.jpg')
gray =  cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray,(3,3),0)
```

```python
edged1 = cv2.Canny(blurred,10,100)
edged2  =  cv2.Canny(blurred,30,150)
edged3  =  cv2.Canny(blurred,240,250)
# 30,150 # mid threshold
# 240,250 # tight threshold
```

```python
cv2.imshow('Original image ',image)
cv2.imshow('Wide threshold Edged image',edged1)
cv2.imshow('Mid threshold Edged image',edged2)
cv2.imshow('Tight threshold Edged image',edged3)
cv2.waitKey(0)
```

```python
contours,_ =  cv2.findContours(edged1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
image_copy = image.copy()

cv2.drawContours(image_copy,contours,-1,(0,255,0),2)
print(len(contours),"objects were found in this image.")

cv2.imshow('Edged  image ',edged1)
cv2.imshow('contours image',image_copy)
cv2.waitKey(0)
```
